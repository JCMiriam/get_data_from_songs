```python
import pandas as pd
import os

# Ruta del archivo CSV grande
file_path = "data/genius_songs.csv"

# Tamaño de cada parte (en número de filas)
rows_per_chunk = 100000  # Puedes ajustar este valor según tus necesidades

# Carpeta para guardar los archivos divididos
output_folder = "data/split_files"
os.makedirs(output_folder, exist_ok=True)

# Contador para los archivos divididos
file_counter = 1

# Leer y dividir el archivo CSV en partes
try:
    for chunk in pd.read_csv(file_path, chunksize=rows_per_chunk):
        output_file = os.path.join(output_folder, f"genius_songs_part_{file_counter}.csv")
        chunk.to_csv(output_file, index=False)
        print(f"Parte {file_counter} guardada: {output_file}")
        file_counter += 1

    print("División completada.")
except Exception as e:
    print(f"Error al dividir el archivo: {e}")
```
